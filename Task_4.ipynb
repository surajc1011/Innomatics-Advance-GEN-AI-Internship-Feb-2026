```python
#Problem 1: Social Media – Post Engagement Analyzer
def post_engagement_analyzer(likes_list):
    total_likes = 0

    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like

    print("Total Likes:", total_likes)

    # Check status
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example Usage
likes = [200, 150, 300, 400]
post_engagement_analyzer(likes)
```

```python
#Problem 3: Agriculture – Rainfall Adequacy Checker
def rainfall_checker(rainfall_data, required_level):
    total_rainfall = 0

    # Loop to calculate total rainfall
    for rain in rainfall_data:
        total_rainfall += rain

    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_data)

    print("Average Rainfall:", int(average_rainfall))

    # Check adequacy
    if average_rainfall >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example Usage
season_rainfall = [70, 75, 80, 63]
required = 70
rainfall_checker(season_rainfall, required)
```

```python
#Problem 4: Social Media – Duplicate Account Detection
def duplicate_account_detection(usernames):

    unique_usernames = set(usernames)

    # Check for duplicates
    if len(unique_usernames) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example Usage
users = ["user1", "user2", "user3", "user1"]
duplicate_account_detection(users)
```

```python
#Problem 6: Agriculture – Premium Crop Price Filter
def premium_crop_filter(prices):

    premium_crops = []

    # Filter prices above ₹2000
    for price in prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)


# Example Usage
crop_prices = [1500, 2500, 1800, 3200, 2000]
premium_crop_filter(crop_prices)
```

```python
#Problem 8: Banking – Daily Transaction Limit Checker
def transaction_limit_checker(amount):

    daily_limit = 50000

    print("Transaction Amount:", amount)

    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


# Example Usage
transaction_limit_checker(60000)
```

```python
#Problem 9: E-Learning – Student Attendance Eligibility System
def attendance_eligibility(attendance_list):

    total_classes = len(attendance_list)
    attended_classes = 0

    # Loop to count present days
    for status in attendance_list:
        if status == "P":
            attended_classes += 1

    # Calculate attendance percentage
    percentage = (attended_classes / total_classes) * 100

    print("Attendance Percentage:", percentage)

    # Check eligibility
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example Usage
attendance = ["P", "P", "A", "P", "P"]  # 4 out of 5 present
attendance_eligibility(attendance)
```

```python
#Problem 10: Smart Electricity Bill Analyzer
def electricity_bill_analyzer(units):

    bill = 0

    # Billing Slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return bill, usage_status


# Example Usage
units_consumed = 250
total_bill, status = electricity_bill_analyzer(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Level:", status)
```
