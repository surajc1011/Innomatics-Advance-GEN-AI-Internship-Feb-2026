```python
#Problem Statement 1:Smart Parking Lot Management System

def smart_parking(capacity, logs):
    current_vehicles = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT" and current_vehicles > 0:
            current_vehicles -= 1

        # Track peak usage
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    print("Currently Parked Vehicles:", current_vehicles)
    print("Peak Parking Usage:", peak_usage)

    if current_vehicles > capacity:
        print("Parking Status: Exceeded Capacity!")
    else:
        print("Parking Status: Available")


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking(capacity, vehicle_logs)
```

```python
# Problem Statement 2:Online Food Delivery Time Estimator
def estimate_delivery_time(distance, traffic, weather):
    # Base time calculation (Assume 5 minutes per km)
    eta = distance * 5

    # Traffic delay
    if traffic == "Low":
        eta += 5
    elif traffic == "Medium":
        eta += 10
    elif traffic == "High":
        eta += 15

    # Weather delay
    if weather == "Clear":
        eta += 0
    elif weather == "Cloudy":
        eta += 5
    elif weather == "Rainy":
        eta += 10
    elif weather == "Stormy":
        eta += 20

    print("Estimated Delivery Time:", eta, "minutes")


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic, weather)
```

```python
# Problem Statement 3:Movie Theatre Seat Occupancy Analyzer
def seat_occupancy(total_seats, booked_seats):
    booked_count = 0

    # Count booked seats using loop
    for seat in booked_seats:
        if seat == 1:
            booked_count += 1

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy), "%")

    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider opening additional show")
    else:
        print("Show Status: Seats Available")


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

seat_occupancy(total_seats, booked_seats)
```

```python
# Problem Statement 4:Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    total = 0

    # Calculate total using loop
    for reading in cpu_readings:
        total += reading

    # Calculate average
    average = total / len(cpu_readings)

    print("Average CPU Load:", int(average), "%")

    # Classification
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

```python
# Problem Statement 5:Smart Classroom Resource Usage Monitor
def classroom_resource_monitor(resource_usage):
    overused = []

    # Define usage limit (example: 8 hours per day)
    limit = 8

    # Loop through dictionary
    for resource, hours in resource_usage.items():
        if hours > limit:
            overused.append(resource)

    # Display results
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(resource_usage)
```

```python
# Problem Statement 6: Online Event Registration Capacity Controller
def event_registration_controller(capacity, registrations):
    confirmed = 0
    waitlisted = 0

    # Loop through registrations count
    for i in range(registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if waitlisted > 0:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
capacity = 100
registrations = 105

event_registration_controller(capacity, registrations)
```
